## Lets see some of the insights from this dataset 

-- selling price of houses <br>
-- sale types <br>
-- sale conditions<br>
-- how old are properties<br>


```python
# selling priice of the houses

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,15))
sns.histplot(df["SalePrice"], bins=100)
plt.locator_params(axis="x", nbins=20)
plt.show()
```

#### lets see some of the outliers sold more than $350,000 and less than $70,000

```python
outliers = df[(df["SalePrice"] > 350000) | (df["SalePrice"] < 70000)]

outliers.sort_values(by="SalePrice", ascending=False)
```

#### we can see most houses fall between range of ~100,000 - ~200,000 sale price

```python
# lets see what type of sale types were there

plt.figure(figsize=(15,15))
sns.histplot(df["SaleType"])
plt.show()
```

#### we can see most of the houses were sold as secondhand or used following with new homes and court foreclosures

#### lets see what was the average prices for these type of homes

```python
sale_type_prices = df.groupby("SaleType", as_index=False)["SalePrice"].mean().sort_values(ascending=False, by="SalePrice")

plt.figure(figsize=(15,15))
sns.barplot(data=sale_type_prices, x="SaleType", y="SalePrice")
plt.show()
```

#### now lets see the distribution of the age of the houses whenthey were sold
-- from this we will know the average possetion time of homeowners <br>
-- later we will also see the years in which houses were sold most.

```python
df["avg_poss_year"] = df["YrSold"] - df["YearBuilt"]

plt.figure(figsize=(15,15))
sns.histplot(df["avg_poss_year"], bins=50)
plt.show()
```
